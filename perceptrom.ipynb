```python
import numpy as np
import matplotlib.pyplot as  plt
from matplotlib.animation import FuncAnimation
```

```python
positive_points=np.array([(-2,3),(0,1),(2,-1)])
negative_points=np.array([(-2,1),(0,-1),(2,-3)])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(positive_points[:,0],positive_points[:,1], color='blue',marker='o',label='positive calss')
plt.scatter(negative_points[:,0],negative_points[:,1], color='red',marker='x',label='negative calss')
plt.legend(loc='best')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('initial data points')
plt.grid('True')
plt.show
```

```python
X=np.vstack((positive_points,negative_points))
y=np.array([1]*len(positive_points)+[-1]*len(negative_points))
```

```python
def perceptron(X,y,steps=10):
    theta=np.zeros(X.shape[1])
    theta0=0
    for t in range(steps):
        for i,xi in enumerate(X):
            if y[i]*(np.dot(xi,theta)+theta0) <=0:
                theta+=y[i]*xi
                theta0 += y[i]
                break
        plot_decesion_boundary(theta,theta0,t+1)
    return theta,theta0    
```

```python
def plot_decesion_boundary(theta,theta0,iteration):
    plt.figure(figsize=(8,6))
    plt.scatter(positive_points[:,0],positive_points[:,1], color='blue',marker='o',label='positive calss')
    plt.scatter(negative_points[:,0],negative_points[:,1], color='red',marker='x',label='negative calss')

    x_values=np.linspace(-3,3,2)

    if theta[1] !=0:
        y_values=-(theta[0] * x_values+theta0)/ theta[1]
        plt.plot(x_values,y_values,label=f'Iteration{iteration}')
    else:
        x_val = -theta0 /theta[0]
        plt.axvline(x=x_val,label=f'Iteration{iteration}')

    plt.xlim(-3,3)
    plt.ylim(-4,4)
    plt.xlabel('x1')
    plt.ylabel('y1')
    plt.title(f'decesion boundary at iteration {iteration} ')
    plt.legend()
    plt.grid('True')
    plt.show    
```

```python
theta_final,theta0_final = perceptron(X,y,10)
```
